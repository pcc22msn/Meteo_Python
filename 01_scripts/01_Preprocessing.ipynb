```python
#importamos librerias pertinentes
import pandas as pd
import numpy as np
import datetime
import calendar
```

```python
DF_SBJ = pd.read_csv("C:/Users/PedroCahuana/Desktop/DF_CONTA/DF_SBJ.csv")
DF_SBJ.head() #evaluamos nuestro DF crudo
```

```python
DF_SBJ.dtypes #observamos el tipo de objeto para cada col del DF
```

```python
DF_SBJ.iloc[:,2:7] = DF_SBJ.iloc[:,2:7].replace("S/D", np.nan) #reemplazamos S/D por valores nulos numpy (NaN)
DF_SBJ.iloc[:,2:7]
```

```python
#cambiamos el tipo de objeto de los parámetros

DF_SBJ["TEMPERATURA MEDIA (°C)"] = DF_SBJ["TEMPERATURA MEDIA (°C)"].astype("float")
DF_SBJ["HUMEDAD RELATIVA (%)"] = DF_SBJ["HUMEDAD RELATIVA (%)"].astype("float")
DF_SBJ["PRECIPITACION (mm)"] = DF_SBJ["PRECIPITACION (mm)"].astype("float")
DF_SBJ["VELOCIDAD_VIENTO (m/s)"] = DF_SBJ["VELOCIDAD_VIENTO (m/s)"].astype("float")
DF_SBJ["DIRECCIÓN_VIENTO (°)"] = DF_SBJ["DIRECCIÓN_VIENTO (°)"].astype("float")
```

```python
DF_SBJ.dtypes #observamos los nuevos tipos de objeto para cada col del DF
```

```python
DF_SBJ["FECHA"] = DF_SBJ["FECHA"].str.replace(" 0:00", "") #quitamos las horas 00:00 que aparecen en la columna fecha
DF_SBJ["FECHA"]
```

```python
DF_SBJ["DATE"] = DF_SBJ["FECHA"].map(str) + " " + DF_SBJ["HORA"] #concatenamos las columnas fecha y hora en otra llamada "date"
DF_SBJ["DATE"]
```

```python
DF_SBJ["DATE"] = pd.to_datetime(DF_SBJ["DATE"], format="%m/%d/%Y %H:%M:%S") #convertimos a datetime la columna "DATE"
DF_SBJ["DATE"]
```

```python
DF_SBJ_filter = DF_SBJ.iloc[:,2:8].set_index("DATE") #creamos un nuevo DF con valores filtrados
DF_SBJ_filter
```

```python
DF_SBJ_filter.dtypes #vemos los tipos de objeto del nuevo DF, deberían ser float
```

```python
#creamos un range time acorde al DF
DF_range_time = pd.date_range("2015-01-01", periods=60, freq="1M")
DF_range_time
```

```python
#rellenamos el DF con fechas faltantes y sus respectivos valores nulos (NAN) - DESDE UN RANGO DEFINIDO CON MI Y MAX (2015-2019)

ultima_fecha = max(DF_range_time)
ultimo_dia_mes = calendar.monthrange(ultima_fecha.year, ultima_fecha.month)[1]
inicio = min(DF_range_time).replace(day=1, hour=0, minute=0, second=0)
fin = ultima_fecha.replace(day=ultimo_dia_mes, hour=23, minute=0, second=0)

index = pd.date_range(start=inicio, end=fin, freq="1h")
```

```python
#observamos y guardamos el nuevo data frame llenado con valores nulos
DF_SBJ_fill = DF_SBJ_filter.reindex(index)
DF_SBJ_fill
```
